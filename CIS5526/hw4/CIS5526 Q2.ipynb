```python
# derivative of the orginal function
fun = lambda x: np.cos(x) + 0.3
def gradDescent2(x0,alpha,iteration):
    currX1 = x0
    bound = 0.000001
    for _ in range(iteration): # iterations until the loop ends
        prevTemp1 = currX1 # store previous x1 to check bounds
        currX1 = prevTemp1 - alpha * fun(prevTemp1) # gradient descent
        if (abs(currX1-prevTemp1)<=bound): # check bound
            break
    return currX1
```

```python
def gradErrors(x,iteration,alphas,alphaErrors,tar):
    i=0
    for j in alphas:
        res = gradDescent2(x,j,iteration)
        minOfGrad = min(tar, key=lambda x:abs(x-res)) # iterate using key to find the closest max or mini
        print(f'alpha:{j}')
        error = abs(abs(minOfGrad)-abs(res))
        print(f'error:{error}')
        alphaErrors[j] = error
        i+=1
    return alphaErrors
```

```python
iteration = 1000
alphas = [0.00001,0.0001,0.001,0.01,0.1,1,10]
minimum = 1.87549
nextMiniMax = 2.5332
x0 = [-5,-3,0,3,5] 
errorRate = []


intervalData = []
for j in range(100): 
    # this will generate the data points that goes max and minimum of the function
    t = minimum + (nextMiniMax*j)
    intervalData.extend([t,-1*t])#get the negative and the positive x axis of f

for i in x0:
    alphaErrors=dict() # appends the errors to a dictionary
    print(f'x:{i}')
    gradErrors(i,iteration,alphas,alphaErrors,intervalData)# get errors from GD
    errorRate.append(alphaErrors)#append the errors 
        
```

```python
a=list(errorRate[0].values()) # convert to list to plot
print(a)
# plt.ylim(0.0001, 2)
# plt.xlim(0.0001, 1.2)
plt.title(f'when X={x0[0]} Alphas by Error Rate')
plt.xlabel('alphas')
plt.ylabel('Error Rate')
plt.scatter(alphas[:6],a[:6])
```
